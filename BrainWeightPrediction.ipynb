```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('headbrain.csv')
dataset.head()
```

Considering " x " and " y " values from dataset

```python
X = dataset['Head Size(cm^3)']
Y = dataset['Brain Weight(grams)']
```

```python
plt.figure(figsize=(6,4), dpi=100)
plt.scatter(X, Y, marker='o', color='r')
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

# calculating slope (m) 
slope_b1 = sum((X - mean_x)*(Y - mean_y))/sum(pow(X - mean_x, 2))
intercept_b0 = mean_y - (slope_b1*mean_x)

print(slope_b1, intercept_b0)
```

```python
sample_x = np.arange(np.min(X) - 100, np.max(X) + 100, 1000) 
print((sample_x))
```

```python
expected_y = intercept_b0 + (slope_b1*sample_x)
print(expected_y)
```

```python
plt.figure(figsize=(6,4), dpi=100)
plt.scatter(X, Y, marker='o', color='r', alpha=0.7)
plt.plot(sample_x, expected_y, color='g')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(dataset[['Head Size(cm^3)']], dataset[['Brain Weight(grams)']])
```

```python
intercept_b0 = model.intercept_[0]
slope_b1 = model.coef_[0]
expected_y = intercept_b0 + (slope_b1*sample_x)
```

```python
expected_y = expected_y
```
